```python
from geopy.geocoders import Nominatim
import time

# Defining a function to geocode the user location of each tweet
def geocode_location(location):
    try:
        geolocator = Nominatim(user_agent="krishnakamalbaishnab")
        time.sleep(1)
        location = geolocator.geocode(location)
        if location is not None:
            return location.latitude, location.longitude
        else:
            return None
    except:
        return None

# Applying the geocode_location function to the 'user_location' column of the DataFrame
CovidTweets500['lat_long'] = CovidTweets500['user_location'].apply(geocode_location)

# Spliting the 'lat_long' column into two separate columns for latitude and longitude
CovidTweets500[['latitude', 'longitude']] = pd.DataFrame(CovidTweets500['lat_long'].tolist(), index=CovidTweets500.index)
```

The "geocode_location" function takes a location string as input and uses the Nominatim geocoding service to retrieve the latitude and longitude coordinates for that location. The function also includes a 1-second delay between geocoding requests to avoid exceeding the usage limits of the geocoding service.

The "apply" method is used to apply the "geocode_location" function to the "user_location" column of the DataFrame. The resulting geocoded latitude and longitude coordinates are stored in a new "lat_long" column of the DataFrame.

Finally, the "lat_long" column is split into two separate columns for latitude and longitude, using the "pd.DataFrame" method to convert the "lat_long" column to a DataFrame, and the "tolist" method to convert the DataFrame to a list. The resulting latitude and longitude values are stored in new "latitude" and "longitude" columns of the DataFrame.
